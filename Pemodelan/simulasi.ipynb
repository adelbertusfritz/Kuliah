```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ocean grid dimensions
grid_size = 100
seabed_displacement = np.zeros((grid_size, grid_size))

# Simulating a Gaussian displacement (simplified)
x, y = np.meshgrid(np.linspace(-1, 1, grid_size), np.linspace(-1, 1, grid_size))
seabed_displacement = np.exp(-5 * (x**2 + y**2))  # Simulates initial displacement

plt.imshow(seabed_displacement, cmap='coolwarm')
plt.colorbar()
plt.title('Seabed Displacement')
plt.show()
```

```python
# Time step and grid step parameters
dt = 0.01  # Time step
dx = 1.0   # Space step
g = 9.81   # Gravitational constant

# Initial wave height and velocity (simplified)
wave_height = np.zeros(grid_size)
wave_velocity = np.zeros(grid_size)

# Initial conditions
wave_height[50] = 1.0  # Disturbance at the center

# Propagation loop (simplified)
for t in range(100):
    # Finite difference approximation of shallow water equations
    for i in range(1, grid_size - 1):
        wave_velocity[i] += dt * (-g * (wave_height[i] - wave_height[i-1]) / dx)
    for i in range(1, grid_size - 1):
        wave_height[i] += dt * (-wave_velocity[i] * (wave_height[i] - wave_height[i-1]) / dx)

plt.plot(wave_height)
plt.title('Wave Propagation')
plt.show()
```
